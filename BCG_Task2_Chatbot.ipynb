```python
import pandas as pd


df = pd.read_csv('/content/financial_data.csv')

def financial_chatbot():
    print("🤖 Welcome to BCG Financial Insights Chatbot!")
    print("Ask me anything about Microsoft, Apple, or Tesla (type 'exit' to quit)\n")

    while True:
        query = input("You: ").strip().lower()

        if query == 'exit':
            print("👋 Thank you for using BCG Financial Chatbot!")
            break


        if 'revenue' in query and 'growth' in query:
            growth = df.groupby('Company')['Total Revenue'].pct_change() * 100
            print("Bot:", growth.dropna().round(2))

        elif 'microsoft' in query or 'apple' in query or 'tesla' in query:
            company = [c for c in ['Microsoft', 'Apple', 'Tesla'] if c.lower() in query][0]
            data = df[df['Company'] == company]
            print(f"Bot: Here is data for {company}:\n", data)

        elif 'compare' in query or 'which' in query:
            print("Bot: Microsoft showed strongest revenue growth. Apple stable. Tesla slight decline in 2025.")

        else:
            print("Bot: I can answer about Revenue, Net Income, Growth, or compare companies. Try: 'Microsoft revenue 2025'")

financial_chatbot()
```

```python
from IPython.display import HTML, display

html_code = """
<!DOCTYPE html>
<html>
<head>
    <title>BCG Financial Chatbot</title>
    <style> body {font-family: Arial;} .chat {border:1px solid #ccc; padding:10px; height:300px; overflow-y:scroll;} </style>
</head>
<body>
    <h2>BCG GenAI Financial Chatbot Prototype</h2>
    <div class="chat" id="chat"></div>
    <input type="text" id="userInput" placeholder="Ask about revenue, growth...">
    <button onclick="sendMessage()">Send</button>

    <script>
        function sendMessage() {
            let input = document.getElementById('userInput').value;
            let chat = document.getElementById('chat');
            chat.innerHTML += '<p><strong>You:</strong> ' + input + '</p>';

            // Simple responses
            let response = "I'm a prototype. Ask about Microsoft revenue or growth.";
            if (input.toLowerCase().includes('revenue')) response = "Microsoft 2025 Revenue: $281,724 million";
            if (input.toLowerCase().includes('growth')) response = "Microsoft has ~15% revenue growth.";

            chat.innerHTML += '<p><strong>Bot:</strong> ' + response + '</p>';
            document.getElementById('userInput').value = '';
            chat.scrollTop = chat.scrollHeight;
        }
    </script>
</body>
</html>
"""

display(HTML(html_code))
```
